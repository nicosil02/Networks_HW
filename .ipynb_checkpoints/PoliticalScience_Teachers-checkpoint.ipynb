```python
import pandas as pd
LinkToData="https://docs.google.com/spreadsheets/d/e/2PACX-1vSj-s_vkNgQb0zgICQcazubHeEju2g9sQ16qeRU5XmY35spPF49JrzdX82rXGpLaRpDbWlBfgjybRfi/pub?gid=1202256999&single=true&output=csv"
SocialesData = pd.read_csv(LinkToData,
                          index_col=0) 
```

```python
SocialesData.head()
```

Transforamos los valores a integrer y corregimos NAs que deberían ser 0.

```python
SocialesData = SocialesData.fillna(0).astype(int)
SocialesData.columns = SocialesData.columns.str.strip()
SocialesData.index = SocialesData.index.map(str.strip)
```

```python
SocialesData.columns #Revisamos las columnas
```

### Adjacency

```python
varsToDrop=["Principal","Asociado","Auxiliar","Contratado"]
adjacency=SocialesData.drop(varsToDrop,axis=1)

#result
adjacency
```

```python
import networkx as nx
TeachersNet = nx.from_pandas_adjacency(adjacency)
```

```python
TeachersNet.remove_edges_from(nx.selfloop_edges(TeachersNet)) #We take away self-relationships
```

### Añadimos categorías de docentes

```python
dict(zip(SocialesData.index,SocialesData['Principal']))
dict(zip(SocialesData.index,SocialesData['Asociado']))
dict(zip(SocialesData.index,SocialesData['Auxiliar']))
dict(zip(SocialesData.index,SocialesData['Contratado']))



```

```python
attributeToAdd_1=dict(zip(SocialesData.index,SocialesData['Principal']))
attributeToAdd_2=dict(zip(SocialesData.index,SocialesData['Asociado']))
attributeToAdd_3=dict(zip(SocialesData.index,SocialesData['Auxiliar']))
attributeToAdd_4=dict(zip(SocialesData.index,SocialesData['Contratado']))

nx.set_node_attributes(TeachersNet, attributeToAdd_1, 'Principal')
nx.set_node_attributes(TeachersNet, attributeToAdd_2, 'Asociado')
nx.set_node_attributes(TeachersNet, attributeToAdd_3, 'Auxiliar')
nx.set_node_attributes(TeachersNet, attributeToAdd_4, 'Contratado')
```

### Graficamos la red y configuramos colores para cada categoría

```python
import matplotlib.pyplot as plt


colors = {
    node: ('blue' if data.get('Principal', 0) == 1 else
           'green' if data.get('Asociado', 0) == 1 else
           'red' if data.get('Auxiliar', 0) == 1 else
           'orange' if data.get('Contratado', 0) == 1 else
           'grey')  # in case one of our categories doesn't fit 
    for node, data in TeachersNet.nodes(data=True)
}


nx.set_node_attributes(TeachersNet, colors, 'color')


plt.figure(figsize=(12, 14))


pos = nx.spring_layout(TeachersNet, k=0.3, iterations=50)  

# Drawing
nx.draw_networkx_nodes(TeachersNet, pos, node_size=700, node_color=[data['color'] for _, data in TeachersNet.nodes(data=True)], alpha=0.9)

# Names
nx.draw_networkx_labels(TeachersNet, pos, font_size=9, font_color='black')

nx.draw_networkx_edges(TeachersNet, pos, width=2, alpha=0.5)

# Guide with the names
from matplotlib.lines import Line2D
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Principal', markersize=10, markerfacecolor='blue'),
    Line2D([0], [0], marker='o', color='w', label='Asociado', markersize=10, markerfacecolor='green'),
    Line2D([0], [0], marker='o', color='w', label='Auxiliar', markersize=10, markerfacecolor='red'),
    Line2D([0], [0], marker='o', color='w', label='Contratado', markersize=10, markerfacecolor='orange')
]
plt.legend(handles=legend_elements, loc='upper right', title="Categorías")

plt.axis('off')  
plt.show()
```

```python
#Revisamos la conectividad entre cada categoría

categorias = ['Principal', 'Asociado', 'Auxiliar', 'Contratado']
conectividad = {}

for categoria in categorias:
    nodos_categoria = [n for n, attr in TeachersNet.nodes(data=True) if attr.get(categoria, 0) == 1]
    subgrafo = TeachersNet.subgraph(nodos_categoria)
    conectividad[categoria] = nx.number_of_edges(subgrafo)  

for cat, conn in conectividad.items():
    print(f"Conectividad interna de {cat}: {conn}")
```

### Exploration

```python
len(TeachersNet.nodes),len(TeachersNet.edges)
```

- 95 nodos: Hay un total de 95 docentes de la Sección de Ciencia Política del Departamento de Ciencias Sociales PUCP

- 27 aristas: Sin embargo, solo hay 27 aristas. Esto representa una reducida relación entre docentes en lo que refiere a producción académica.

```python
len(list(nx.connected_components(TeachersNet))) #Conectividad
```

- Hay 73 componentes conectados en el grafo. En ese sentido, el grafo se puede dividir en 73 subgrupos distintos donde cada subgrupo está formado por nodos que están conectados directamente entre sí, pero no tienen conexiones directas con nodos en otros subgrupos. Esto sugiere la existencia de clusters en la red que se encuentran relativamente aislados entre sí.


```python
list(nx.isolates(TeachersNet)) #nodos aislados
```

```python
#just for the largest component
largest_cc_nodes = max(nx.connected_components(TeachersNet), key=len)
TeachersNet_largest=TeachersNet.subgraph(largest_cc_nodes)

nx.diameter(TeachersNet_largest)
```

- El diámetro del componente más grande es 3, lo que indica que la máxima distancia entre cualquier par de nodos dentro de este componente es de tres enlaces.

```python
# geodesics
geodesics = dict(nx.all_pairs_shortest_path_length(TeachersNet))

# as matrix
MatrixGeo = [[geodesics.get(n,{}).get(m)  for m in TeachersNet.nodes] for n in TeachersNet.nodes]

# as DF (pandas)
MatrixGeo_DF=pd.DataFrame(MatrixGeo,index=TeachersNet.nodes, columns=TeachersNet.nodes)

# we get
MatrixGeo_DF
```

- Los valores en la tabla son las distancias geodésicas, donde cada celda muestra la mínima cantidad de pasos requeridos para ir de un nodo a otro en la red de docentes. Como se aprecia en la mayoría de casos, no existe conexión entre los pares de nodos.

```python
import numpy as np

pd.DataFrame(np.unique(MatrixGeo_DF.values, return_counts=True),index=['valor','freq']).T.set_index('valor')
```

- Se observa que hay 95 nodos autoconectados, 54 conexiones directas y un número decreciente de conexiones que requieren más pasos, con 74, 38 y 6 conexiones para dos, tres y cuatro pasos respectivamente. No obstante, 8758 pares de nodos no tienen conexión directa o indirecta, destacando una significativa falta de cohesión en la red.

```python
MatrixReach_DF=(MatrixGeo_DF>0).replace({True:1,False:0})
MatrixReach_DF
```

- Se mantiene el patrón de falta de conexión entre nodos.

```python
# connectedness
nNodes=len(TeachersNet.nodes)
MatrixReach_DF.sum().sum()/(nNodes*(nNodes-1))
```

- El valor de 0.0192 indica que, en promedio, solo el 1.92% de los pares de nodos están conectados directamente o indirectamente dentro de esta red, reflejando un nivel bajo de conectividad total. 

```python
# geodesics
geodesics_Lcc = dict(nx.all_pairs_shortest_path_length(TeachersNet_largest))

# as matrix
MatrixGeo_Lcc = [[geodesics_Lcc.get(n,{}).get(m)  for m in TeachersNet_largest.nodes] for n in TeachersNet_largest.nodes]

# as DF (pandas)
MatrixGeo_DF_Lcc=pd.DataFrame(MatrixGeo_Lcc,index=TeachersNet_largest.nodes, columns=TeachersNet_largest.nodes)

# we get
MatrixGeo_DF_Lcc
```


- La matriz refleja las distancias geodésicas en el componente más grande de la red, mostrando el número mínimo de enlaces necesarios para conectar cualquier par de nodos.

```python
MatrixReach_DF_Lcc=(MatrixGeo_DF_Lcc>0).replace({True:1,False:0})
MatrixReach_DF_Lcc
```

- En este caso, el valor de 1 indica que hay una conexión directa entre 2 nodos, y el 0, que no existe.

```python
nNodes=len(TeachersNet_largest.nodes)
MatrixReach_DF_Lcc.sum().sum()/(nNodes*(nNodes-1))
```

El valor de conectividad de 1.0 para el componente más grande de la red implica que todos los nodos dentro de esta componente están directamente conectados entre sí, alcanzando la máxima densidad posible.

```python
nx.density(TeachersNet_largest)
```

- La densidad del componente más grande de la red indica que aproximadamente el 24.44% de todas las conexiones posibles entre los nodos se encuentra en este componente. 

```python
actorMeasures=pd.concat([pd.DataFrame.from_dict(nx.degree_centrality(TeachersNet_largest),
                                  orient='index',
                                  columns=['Degree']),
           pd.DataFrame.from_dict(nx.closeness_centrality(TeachersNet_largest),
                                  orient='index',
                                  columns=['Closeness']),
           pd.DataFrame.from_dict(nx.betweenness_centrality(TeachersNet_largest),
                                  orient='index',
                                  columns=['Betweenness'])],axis=1)
## data
actorMeasures
```

- Grado (Degree): Representa la cantidad de conexiones directas que tiene cada actor. Por ejemplo, Javier Ramírez tiene una centralidad de grado del 11.11%, indicando que está directamente conectado con una fracción significativa de la red.

- Cercanía (Closeness): Esta medida indica la distancia promedio de un nodo a todos los otros nodos en la red. Un valor más alto implica una menor distancia promedio. Por ejemplo, Oscar Vidarte, con un valor de 0.8182, puede alcanzar a otros nodos más rápidamente en comparación con otros.

- Intermediación (Betweenness): Muestra la frecuencia con la que un nodo actúa como un puente a lo largo del camino más corto entre otros dos nodos. Por ejemplo, Oscar Vidarte tiene una centralidad de intermediación de 0.8055, lo que lo posiciona como un punto crítico para la transmisión de información a través de la red.

En este apartado se evidencia que el componente más grande se encuentra compuesto por conexiones entre profesores que dictan cursos de Relaciones Internacionales en la carrera de Ciencia Política.

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,10)) #Graficamos

ax.scatter(actorMeasures.Closeness, actorMeasures.Degree,s=1000*actorMeasures.Betweenness,
           c=100*actorMeasures.Betweenness,
           cmap=plt.get_cmap('YlOrRd'), alpha=0.6)

valsForAnnotate=zip(actorMeasures.index,actorMeasures.Closeness,actorMeasures.Degree)
for name,coordX,coordY in valsForAnnotate:
    ax.annotate(name, (coordX,coordY),alpha=0.5)

plt.title("Centrality  (size and  color for Betweenness)", y=-0.1)
plt.xlabel("Closeness")
plt.ylabel("Degree")
plt.show()
```

### Conclusiones:

- Este análisis revela que, contrariamente a lo esperado, no necesariamente los docentes Principales, quienes poseen mayor antigüedad, forman las redes más fuertes de producción académica conjunta. Aunque estos docentes se encuentran en los componentes más grandes y mantienen colaboraciones académicas, las redes más amplias involucran principalmente a docentes que imparten cursos de Relaciones Internacionales. En particular, el profesor Principal Oscar Vidarte se destaca como un nodo central con una alta centralidad de grado del 77.8%. 

    Los docentes Principales suelen participar en al menos una colaboración académica, mientras que los docentes Contratados, generalmente más nuevos en la comunidad, muestran conexiones limitadas. Los docentes Asociados y Auxiliares, aunque conectados, representan un grupo menor y, por lo tanto, no contribuyen significativamente al total de conexiones de la red.

    A pesar de algunas conexiones activas, la red en general exhibe una conectividad limitada, con solo el 1.92% de los pares de nodos conectados directa o indirectamente. Esto sugiere la necesidad de fortalecer las redes de colaboración académica en la universidad, en un esfuerzo por construir una comunidad académica más robusta en la especialidad.
